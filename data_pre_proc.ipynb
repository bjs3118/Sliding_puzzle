```python
"""
Block to convert HEIC to JPG
"""


import os
from PIL import Image
import pyheif

def convert_heic_to_jpg(heic_path, jpg_path):
    heif_file = pyheif.read(heic_path)
    image = Image.frombytes(
        heif_file.mode,
        heif_file.size,
        heif_file.data,
        "raw",
        heif_file.mode,
        heif_file.stride,
    )
    image.save(jpg_path, format="JPEG")

def convert_all_heic_files(folder_path):
    heic_files = [file for file in os.listdir(folder_path) if file.lower().endswith(".heic")]

    for heic_file in heic_files:
        heic_path = os.path.join(folder_path, heic_file)
        jpg_path = os.path.join(folder_path, os.path.splitext(heic_file)[0] + ".jpg")

        convert_heic_to_jpg(heic_path, jpg_path)
        print(f"Converted: {heic_file} to {os.path.basename(jpg_path)}")

# Provide the path to the folder containing .HEIC files
folder_path = "/content/drive/MyDrive/Project_001/data_raw_001"

# Convert all .HEIC files in the specified folder
convert_all_heic_files(folder_path)
```

```python
"""
Block for moving JPG to another folder
"""

import shutil

def move_jpg_files_to_folder(source_folder, destination_folder):
    jpg_files = [file for file in os.listdir(source_folder) if file.lower().endswith(".jpg")]

    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    for jpg_file in jpg_files:
        source_path = os.path.join(source_folder, jpg_file)
        destination_path = os.path.join(destination_folder, jpg_file)

        shutil.move(source_path, destination_path)
        print(f"Moved: {jpg_file} to {os.path.basename(destination_path)}")

# Provide the path to the folder containing .jpg files
source_folder = "/content/drive/MyDrive/Project_001/data_raw_001"

# Provide the path to the new folder where .jpg files will be moved
destination_folder = "/content/drive/MyDrive/Project_001/data_raw_002"

# Move all .jpg files to the specified destination folder
move_jpg_files_to_folder(source_folder, destination_folder)
```
